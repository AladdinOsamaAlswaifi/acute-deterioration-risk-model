# Notebook 02 - Preprocessing & Target Definition

This notebook prepares features for baseline deterioration alert prediction.

Target: nurse_alert (proxy escalation outcome).

```python
import pandas as pd
# Load dataset
from google.colab import files
uploaded = files.upload()

file_name = list(uploaded.keys())[0]
df = pd.read_csv(file_name)

print(df.shape)
df.head()
```

```python
df["target_alert"] = df["nurse_alert"].astype(int)
df["target_alert"].value_counts()
```

```python
df = pd.get_dummies(df, columns=["oxygen_device"],drop_first=True)

print("New shape:", df.shape)
```
